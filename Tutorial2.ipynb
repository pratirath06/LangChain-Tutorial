```python
import os
from api import GROQ_API_KEY
os.environ["GROQ_API_KEY"]= GROQ_API_KEY
from langchain_groq import ChatGroq
model = ChatGroq(model="llama-3.3-70b-specdec")
```

```python
from langchain_core.messages import SystemMessage,AIMessage,HumanMessage
```

```python
messages = [
    SystemMessage(content="You are a helpful assistant"),
    HumanMessage(content="Tell me about AI Message and Human Message in context of LLMs. And also tell about the flow of these messages.")
]
result = model.invoke(messages)
print(result.content)
```

```python
messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="Who is Current FirstLady?")
]

result = model.invoke(messages)
print(result.content)

result = model.invoke("Who is her husband?")
print(result.content)
```

```python
messages = [
    SystemMessage(content="You are a helpful assistant"),
    HumanMessage(content="Who is Current FirstLady?"),
    AIMessage(content="Jill Biden"),
    HumanMessage(content="Who is her husband?"),
    AIMessage(content="Joe Biden"),
    HumanMessage(content= "Who is owner of Tesla"),
    AIMessage(content="Elon Musk"),
    HumanMessage(content="What was my last question")
    
]

result = model.invoke(messages)
print(result.content)
```

```python
### messages=[]
sys_message = SystemMessage(content="You are a helpful assistant.")
messages.append(sys_message)
while True:
    print("1. Ask a question \n2. Print Chat Conversation \n3. Exit")
    i = int(input("Enter:"))
    if(i==1):
        user = input("Ask your Question")
        messages.append(HumanMessage(content=user))
        result = model.invoke(messages)
        messages.append(AIMessage(content=result.content))
        print(result.content)
    elif(i==2):
        print(messages)
    elif(i==3):
        print("Thank You!")
        break
    else:
        print("Invalid Input, try again")
        
```
